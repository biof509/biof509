## Travelling salesman problem

Last week we covered several different exercises including a solver for a simplified version of the [travelling salesman problem](https://en.wikipedia.org/wiki/Travelling_salesman_problem). Instead of finding a cycle we only find the optimal path, there is no requirement to return to the starting point.

This problem has numerous applications and we may want to use our solution again in other projects. Let's take a look at our solution and then decide how we can make it more useful.

```python
coords = [(0,0), (10,5), (10,10), (5,10), (3,3), (3,7), (12,3), (10,11)]


def distance(coords):
    distance = 0
    for p1, p2 in zip(coords[:-1], coords[1:]):
        distance += ((p2[0] - p1[0]) ** 2 + (p2[1] - p1[1]) ** 2) ** 0.5
    return distance


print(distance(coords))

all_options = []
for option in itertools.permutations(coords, len(coords)):
    all_options.append((option, distance(option)))
all_options.sort(key=lambda x: x[1])

print('Shortest route:', all_options[0])
print('Number of routes considered:', len(all_options))
```

Currently, the only way to reuse this code would be to copy and paste it. Our first improvement would be to convert it into a function.

```python
def distance(coords):
    distance = 0
    for p1, p2 in zip(coords[:-1], coords[1:]):
        distance += ((p2[0] - p1[0]) ** 2 + (p2[1] - p1[1]) ** 2) ** 0.5
    return distance

def find_best_route(coords):
    best_distance = distance(coords)
    best = coords
    for option in itertools.permutations(coords, len(coords)):
        option_distance = distance(option)
        if option_distance < best_distance:
            best = option
            best_distance = option_distance
    return (best_distance, best)
```

```python
coords = [(random.randint(0,1000), random.randint(0,1000)) for i in range(8)]
print(coords)
x = np.array([x[0] for x in coords])
y = np.array([y[1] for y in coords])
plt.plot(x, y, 'r-')

best_distance, best_coords = find_best_route(coords)

best_x = np.array([x[0] for x in best_coords])
best_y = np.array([y[1] for y in best_coords])

plt.plot(best_x, best_y, 'bo-')
plt.show()
```

A significant improvement in speed is possible if instead of looking for the __perfect__ solution, we instead content ourselves with a __good__ solution.

This is an __optimization problem__, in which we are attempting to choose a set of parameters (the order of the points) that minimizes a cost function, the distance travelled. This is a very common task in machine learning: 

* optimizing a model to best describe the available data
* dividing a dataset into clusters to minimize the variance between points in each cluster
* etc

## Simulated Annealing

With both our functions made we can now implement the first of these two algorithms.

```python
def simulated_annealing_optimizer(starting_path, cost_func, new_path_func, start_temp, min_temp, steps):
    """Minimizes the path distance for a collection of coordinates
    
    Utilizes simulated annealing
    
    Parameters
    ----------
    starting_path : list
        The collection of coordinates
    cost_func : function
        The distance metric
    new_path_func : function
        Function to create a new path based on the one supplied
    start_temp : float
        Likelihood of accepting a worse path at the beginning of the optimization
    min_temp : float
        Likelihood of accepting a worse path at the end of the optimization
    steps : integer
        The number of rounds of optimization
        
    Returns
    -------
    out : tuple
        The best path found, the distance of the best path, and a history
        of the step, temperature, and current distance of best path"""
    current_path = starting_path[:] # Make a copy of the supplied points
    current_cost = cost_func(current_path) # Initialize the current_cost variable
    temp_factor = -np.log(start_temp / min_temp) # Used to set the temperature for each step
    history = [] # Variable to log results/progress
    for s in range(0, steps): # 
        temp = start_temp * np.exp(temp_factor * s / steps) # Temperature exponentially decreases
        new_path = new_path_func(current_path) # Generate a new path
        new_cost = cost_func(new_path) # Calculate distance
        if (new_cost < current_cost) or (random.random() <= np.exp(-(new_cost - current_cost)/temp)):
            # If new distance is less than current best accept it
            # If new distance is greater than current best we *may* accept it depending on
            # how much worse it is and the current temperature
            current_path = new_path
            current_cost = new_cost
        record = {'step':s, 'temperature':temp, 'current_cost':current_cost, } # Record results
        history.append(record)
    return (current_path, current_cost, history)
```

The key feature here is how we decide to accept an alternate path.

    temp_factor = -np.log(start_temp / min_temp)

This gives an exponent (a power) such that:

start_temp * e**temp_factor = min_temp

Then, within the for loop:

    temp = start_temp * np.exp(temp_factor * s / steps)
    
For the first step, s=0 so we have np.exp(0). This equals 1 and temp is equal to start_temp.
For the last step, s=steps so we have np.exp(temp_factor) and this scales start_temp down to min_temp.

If the new path is worse whether we accept it not is determined by:

    (random.random() <= np.exp(-(new_cost - current_cost)/temp))
    
The difference in the costs between the previous and current routes is divided by temp and scaled to between 0 and 1 by using the negative value as an exponent.

This is then compared with a random number between 0 and 1 to determine whether to accept the new route.

```python
best_path, best_cost, history = simulated_annealing_optimizer(coords, distance, new_path, 1000, 0.01, 1000)
print(best_cost)
plt.plot([i['current_cost'] for i in history])
plt.show()

plt.plot([i[0] for i in best_path], [i[1] for i in best_path])
plt.show()
```

## Genetic Algorithm

Implementing a genetic algorithm is slightly more challenging

```python
def select_best(population, cost_func, num_to_keep):
    """Select the best candidate paths from a selection of paths
    
    Parameters
    ----------
    population : list
        The candidate paths
    cost_func : function
        The distance metric
    num_to_keep : integer
        The number of paths to return
    
    Returns
    -------
    out : list
        The best paths"""
    # Add the distance of each path to the list
    scored_population = [(i, cost_func(i)) for i in population]
    # Sort by distance
    scored_population.sort(key=lambda x: x[1])
    #Return the number required from the best routes
    return [i[0] for i in scored_population[:num_to_keep]]

def recombine(population):
    """Create a new path from a population of paths
    
    Two paths are selected and recombined together similar to meiosis
    
    Parameters
    ----------
    population : list
        The candidate paths
    Returns
    -------
    out : list
        The new path"""
    # Randomly choose two parents
    options = list(range(len(population)))
    random.shuffle(options)
    partner1 = options[0]
    partner2 = options[1]
    # Choose a split point, take the first parents order to that split point, 
    # then the second parents order for all remaining points
    split_point = random.randint(0, len(population[0])-1)
    child = population[partner1][:split_point]
    # For any point not yet in child add it from partner2
    for point in population[partner2]:
        if point not in child:
            child.append(point)
    return child


def genetic_algorithm_optimizer(starting_path, cost_func, new_path_func, pop_size, generations):
    """Minimizes the path distance for a collection of coordinates
    
    Utilizes a genetic algorithm
    
    Parameters
    ----------
    starting_path : list
        The collection of coordinates
    cost_func : function
        The distance metric
    new_path_func : function
        Function to create a new path based on the one supplied
    pop_size : integer
        The number of paths generated and considered at each generation
    generations : integer
        The number of rounds
    
    Returns
    -------
    out : tuple
        The best path found, the distance of the best path, and a history
        of the generation, and current distance of best path"""
    # Create a starting population by randomly shuffling the points
    # Multiple paths are created to create the population
    # All paths have a random order at this stage
    population = []
    for i in range(pop_size):
        new_path = starting_path[:]
        random.shuffle(new_path)
        population.append(new_path)
    history = [] # Variable to log results/progress
    # Take the top 25% of routes and recombine to create new routes, repeating for generations
    for i in range(generations):
        # Select the top 25% of routes
        pop_best = select_best(population, cost_func, int(pop_size / 4))
        # Recombine these top 25% to return the population size to pop_size
        new_population = []
        for i in range(pop_size):
            new_population.append(recombine(pop_best))
        population = new_population
        record = {'generation':i, 'current_cost':cost_func(population[0]),} # Record results
        history.append(record)
    return (population[0], cost_func(population[0]), history)
```

The key features are:
- Testing multiple possible routes at each round
- Combining the best together
    
The algorithm begins with a completely random ordering. Most of these routes will be very bad. Some routes will have short runs of efficiently ordered points and slightly shorter distances as a result. These routes are carried over to the next population. By combining routes together and then keeping only the best these short runs of efficient ordering should be merged together and maintained. The inefficient ordering will eventually be lost.

```python
best_path, best_cost, history = genetic_algorithm_optimizer(coords, distance, new_path, 500, 100)
print(best_cost)
plt.plot([i['current_cost'] for i in history])
plt.show()        
    
plt.plot([i[0] for i in best_path], [i[1] for i in best_path])
plt.show()
```
